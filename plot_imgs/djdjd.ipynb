```python
import matplotlib.pyplot as plt
import numpy as np

# Data setup
finetuning_epoch = np.arange(1, 6)  # Assuming 5 finetuning epochs
energy_consume = {
    "baseline": np.array([9.79724471, 9.79724471, 9.79724471, 9.79724471, 9.79724471]),
    "MAML": np.array([10.39355328, 10.78571753, 10.6665126 , 10.77705026, 10.60893857]),
    "pre_train": np.array([10.9930054 , 10.99721994, 10.83813681, 10.90271036, 11.01029081]),
    "random": np.array([10.61537241, 10.85323489, 10.67971128, 10.54592625, 10.8606109])
}

# Plotting
plt.figure(figsize=(10, 6))
for label, data in energy_consume.items():
    plt.plot(finetuning_epoch, data, label=label, marker='o')
    
plt.title('Energy Consumption over Finetuning Epochs in 10x5')
plt.xlabel('Finetuning Epoch')
plt.ylabel('Energy Consume')
plt.legend()
# plt.grid(True)
plt.show()
```

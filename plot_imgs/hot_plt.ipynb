```python
import numpy as np

# 创建三个 11x10 的张量
tensor1 = np.random.rand(11, 10)
tensor2 = np.random.rand(11, 10)
tensor3 = np.random.rand(11, 10)

# 提取特定的列：第3列、第5列、第7列和第10列 (索引从0开始)
columns_to_extract = [2, 4, 6, 9]
extracted_columns_1 = tensor1[:, columns_to_extract]
extracted_columns_2 = tensor2[:, columns_to_extract]
extracted_columns_3 = tensor3[:, columns_to_extract]

# 计算这些列的均值
mean_columns_1 = np.mean(extracted_columns_1, axis=0)
mean_columns_2 = np.mean(extracted_columns_2, axis=0)
mean_columns_3 = np.mean(extracted_columns_3, axis=0)

# 比较三个张量在这些列的平均值，以确定相对优势
mean_comparison = np.array([mean_columns_1, mean_columns_2, mean_columns_3])


# 计算张量1和张量2相对于张量3（基准）的差异
difference_1 = mean_columns_1 - mean_columns_3
difference_2 = mean_columns_2 - mean_columns_3

# 计算所有三个张量列均值之间的最大差异
max_difference = np.max(np.abs(mean_comparison - mean_comparison[:, np.newaxis]), axis=(0, 1))

# 为了使优势值落在-1到1之间，将差异值除以最大差异
# 注意：为避免除以0的情况，我们使用np.clip确保最大差异不为0
normalized_advantage_1 = difference_1 / np.clip(max_difference, 1e-10, None)
normalized_advantage_2 = difference_2 / np.clip(max_difference, 1e-10, None)

normalized_advantage_1, normalized_advantage_2

```

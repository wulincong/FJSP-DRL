```python
import matplotlib.pyplot as plt
import numpy as np

# 模拟数据：创建一个随机数据矩阵
data = np.random.rand(10, 10)

def hot_plt(data, title='Algorithm Performance Heatmap', x_label="Number of Jobs", y_label="Number of Machines"):
    labels = [5, 8, 11, 14, 17, 20, 23, 25]

    # 绘制热图
    plt.figure(figsize=(8, 6))
    heatmap = plt.imshow(data, interpolation='nearest', cmap='viridis')
    plt.colorbar(heatmap)
    # 我们假设data的尺寸与labels数组的长度匹配
    plt.xticks(range(len(labels)), labels)
    plt.yticks(range(len(labels)), labels)
    # 添加标题和轴标签
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)

    # 显示图表
    plt.show()
```

```python
model1_list = []
with open("./model1.txt", "r") as f:
    for num in f:
        model1_list.append(float(num))

model1 = np.array(model1_list).reshape(8,8)
print(model1)

model2 = model1+100

hot_plt(model1 / model2)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 定义解集
solutions = np.array([
    [235, 277.85],
    [239, 257.66],
    [265, 242.28],
    [260, 232.29],
    [339, 226.75],
    [1928, 163.19]
])

# 定义Pareto前沿解集
pareto_front = np.array([
    [235, 277.85],
    [239, 257.66],
    [339, 226.75],
    [1928, 163.19]
])

# 参考点（通常选择一个比解集中的所有值更差的点）
reference_point = np.array([2000, 300])

# 绘制解集
plt.scatter(solutions[:, 0], solutions[:, 1], c='blue', label='Solutions')
plt.scatter(pareto_front[:, 0], pareto_front[:, 1], c='red', label='Pareto Front')

# 连线Pareto前沿
plt.plot(pareto_front[:, 0], pareto_front[:, 1], 'r-', label='Pareto Curve')

# 填充超体积区域
plt.fill_between(np.append(pareto_front[:, 0], reference_point[0]),
                 np.append(pareto_front[:, 1], reference_point[1]),
                 reference_point[1], color='gray', alpha=0.3, label='Hypervolume')

# 标注
plt.xlabel('Makespan')
plt.ylabel('Total Energy')
plt.title('Pareto Front and Hypervolume')
plt.legend()

# 显示图形
plt.grid(True)
plt.show()
```
